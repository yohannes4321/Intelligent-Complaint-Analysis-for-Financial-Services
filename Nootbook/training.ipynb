```python
import pandas as pd 
from langchain_community.document_loaders import TextLoader

read=pd.read_csv('../data/filtered_complaints.csv')
```

```python
from langchain_core.documents import Document
document=[]
for _,row in read.iterrows():
    doc=Document(
        page_content=str(row['Consumer complaint narrative']),
        metadata={"Product":row[['Product','Complaint ID']]}
    )
    document.append(doc)

        
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
split=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100)
all_chunks=[]
chunk =split.split_documents(document)
   
```
